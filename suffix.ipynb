```python
names = open("names.txt").read().splitlines()
print(type(names))
```

```python
print(len(names))
print(max(len(w) for w in names))
print(min(len(w) for w in names))
```

```python
#Create a bigram:
bigrams = {}
for name in names:
    augmented_name = ['.'] + list(name) + ["."]
    for i in range(len(augmented_name)):
        bigram = tuple(augmented_name[i:i+2])
        bigrams[bigram] = bigrams.get(bigram,0) + 1
```

```python
#Lets make a dict mapping index to string:
chars = sorted(list(set("".join(names))))
```

```python
itos = {}
for i in range(1,len(chars)+1):
    itos[i] = chars[i-1]
```

```python
itos[0] = '.'
```

```python
stoi = {s:i for i,s in itos.items()}
```

```python
import torch
```

```python
#Now lets create a 2D array with the frequency mappings
N = torch.zeros((27,27),dtype=torch.int32)
```

```python
for name in names:
    augmented_name = ['.'] + list(name) + ['.']
    for char1, char2 in zip(augmented_name,augmented_name[1:]):
        idx1 = stoi[char1]
        idx2 = stoi[char2]
        N[idx1,idx2] +=1
```
